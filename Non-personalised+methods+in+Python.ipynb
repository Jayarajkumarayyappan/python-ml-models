## Introduction to the Data from A Movie Recommender System

MovieLens (100k)
MovieLens is a data set which consists of 100,000 ratings (1-5) from 943 users on 1682 movies.

Each user has rated at least 20 movies.

Simple demographic information for the users includes age, gender, occupation, zip code.

The data was collected through the MovieLens web site (movielens.umn.edu) during the seven-month period from September 19th, 1997 through April 22nd, 1998.

This data has been cleaned up and users who had less than 20 ratings or did not have complete demographic information were removed from this data set.

## Ratings and User Info
Here are the member variables on u dataset.

### u.data

The full `u` data set, 100000 ratings by 943 users on 1682 items.

Each user has rated at least 20 movies. 

Users and items are numbered consecutively from 1. 

The data is randomly ordered.

This is a tab separated list of

- user id | item id | rating | timestamp

- The time stamps are unix seconds since 1/1/1970 UTC  

### u.info

The number of users, items, and ratings in the u data set.

### u.item

Information about the items (movies); this is a tab separated list of

              movie id | movie title | release date | video release date |

              IMDb URL | unknown | Action | Adventure | Animation |

              Children's | Comedy | Crime | Documentary | Drama | Fantasy |

              Film-Noir | Horror | Musical | Mystery | Romance | Sci-Fi |

              Thriller | War | Western |

The last 19 fields are the genres, a 1 indicates the movie is of that genre, a 0 indicates it is not; movies can be in several genres at once.

### u.genre

 

A list of the genres.

The movie ids are the ones used in the u.data data set.

### u.user

Demographic information about the users;

this is a tab separated list of

              user id | age | gender | occupation | zip code

The user ids are the ones used in the u.data data set.

### u.occupation
A list of the occupations.

### Load the rating data to pandas dataframe

```python
names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('u.data', sep='\t', names=names)
df.head()
```

### Load the item data

```python
i_cols = ['movie id', 'movie title' ,'release date','video release date', 'IMDb URL', 'unknown', 'Action', 'Adventure',
 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama', 'Fantasy',
 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western']
items = pd.read_csv('u.item', sep='|', names=i_cols, encoding='latin-1')

items.head()
```

### Convert data into user-item rating matrix and user-item binary rating matrix

#### check the number of users and items

```python
n_users = df.user_id.unique().shape[0]
n_items = df.item_id.unique().shape[0]
print(f"{n_users} users")
print(f"{n_items} items")
```

#### Construct the user-time rating matrix

```python
ratings = np.zeros((n_users, n_items))
```

```python
ratingsNum = np.zeros((n_users, n_items))
```

```python
for row in df.itertuples():
    ratings[row[1]-1, row[2]-1] = row[3]
    ratingsNum[row[1]-1, row[2]-1] = 1
print(ratings)
print(ratingsNum)
```

### Calculate movie’s popularity/average rating

Calculate the popularity: the number of 'rated' times

```python
itemRateNum = ratingsNum.sum(axis=0)
print(itemRateNum)
```

Calculate the total ratings obtained by each item

```python
itemRateSum = ratings.sum(axis=0)
print(itemRateSum)
```

Calculate the average rating for each item

```python
itemRateAvg = itemRateSum/itemRateNum
print(itemRateAvg)
```

### TopPop

#### Specify the number of recommended items

```python
top_n = 5
```

#### Mask the 'rated' items for the active user

```python
activeUser = 0
```

```python
mask_activeUser = ratings[activeUser, :] > 0
```

#### Make a copy of itemRateNum

```python
itemRateNumCurrent = itemRateNum.copy()
```

#### Exclude reated items from recommendation by setting their popularity to 0

```python
itemRateNumCurrent[mask_activeUser] = 0
```

#### Sort Movies by their populairty in descending order. 
The first one is the most popular one

```python
itemSortInd = itemRateNumCurrent.argsort()[::-1]
```

#### Find the top_n items and print

```python
top_ind = itemSortInd[:top_n]
```

```python
items.iloc[top_ind,:][['movie id', 'movie title']]
```
